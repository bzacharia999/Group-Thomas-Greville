```python
#6[C]: Population Maps

#Importing
import numpy as np
from datascience import *
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#Data source: https://data.humdata.org/dataset/worldpop-population-density-for-somalia
df = Table.read_table('som_pd_2020_1km_ASCII_XYZ.csv')

#Random sampling 25% of data if the device is not good enough to run the whole table
df_r = df.sample(187042, with_replacement=False)

#Columns
x = df.column('X')
y = df.column('Y')

#Map
plt.figure(figsize=(8,6))
plt.scatter(x, y, s=1, c= df.column('Z'), cmap='copper_r', vmin=4,vmax=100)
plt.title('Population Density Map of Somalia in 2020')
plt.xlabel('longitude')
plt.ylabel('latitude')

#Bar
bar = plt.colorbar(plt.scatter(x, y, s=1, c= df.column('Z'), cmap='copper_r', vmin=4,vmax=100))
bar.set_label('Population density')

plt.show()
```
