```python
def get_country_population_data(country_code):
    """
    Fetches World Bank data for all indicators in the filtered_df 
    for a specific country.
    """
    SOURCE = 40 # "Population estimates and projections
    indicators = wbdata.get_indicators(source=SOURCE)
    df_indicators = pd.DataFrame(indicators) # returns all of the indicators as a DataFrame containing ids and other information
    pattern = r'^SP\.POP\..*(\.FE|\.MA)$' # the regex pattern matches ids that start with SP.POP and end in either FE or MA
    df_filtered = df_indicators[df_indicators['id'].str.contains(pattern, na=False, regex=True)].copy()
    variable_labels = dict(zip(df_filtered['id'], df_filtered['name'])) # returns dictionary
    df_results = wbdata.get_dataframe(variable_labels, country=country_code)
    
    return df_results
```
